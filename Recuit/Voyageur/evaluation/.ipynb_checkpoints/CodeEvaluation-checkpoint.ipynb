# L’algorithme recuit simulé

L'algorithme du recuit simulé repose sur deux concepts clés de la physique statistique. Tout d'abord, en atteignant l'équilibre thermodynamique à une température T, la probabilité qu'un système physique possède une énergie E est proportionnelle au facteur de Boltzmann, défini comme exp(-E/KT) où K représente la constante de Boltzmann. L'utilisation de l'algorithme de Metropolis est une étape cruciale pour simuler l'évolution d'un système vers son équilibre thermodynamique. En partant d'une configuration initiale, le système subit des modifications élémentaires. Si ces modifications réduisent l'énergie E, elles sont immédiatement acceptées. Cependant, si Δ E > 0, elles sont tout de même acceptées avec une probabilité exp(- Δ E/T), particulièrement à basse température où seules de légères augmentations d'énergie sont permises. La suite des configurations converge ainsi progressivement vers un état d'équilibre thermodynamique à la température T. Il est important de noter que la diminution "légère" de la température facilite cette convergence, évitant ainsi un ralentissement si la température est abaissée trop rapidement. Lorsque la température approche zéro, le système tend vers un minimum d'énergie, signalant sa proximité avec un état d'équilibre thermodynamique optimal.[source](https://www.i2m.univ-amu.fr/~preaux)
